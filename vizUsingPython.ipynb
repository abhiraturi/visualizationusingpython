```python
df_can.columns=list(map(str,df_can.columns))
years=list(map(str,range(1980,2014)))
years
```

```python
#Visualize the top 5 countries that contributed the most immigrants to Canada from 1980 to 2013 using an Area plot
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5 = df_can.head()

# transpose the dataframe
df_top5 = df_top5[years].transpose() 

df_top5.head()
df_top5.index = df_top5.index.map(int) 
df_top5.plot(kind='area', 
             stacked=False,
             figsize=(20, 10), 
             )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
df_top5.plot(kind='area', 
              alpha=0.25,
             stacked=False,
             figsize=(20, 10),
            )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
# option 2: preferred option with more flexibility
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
#Create a stacked area plot of the 5 countries that contributed the least to immigration to Canada from 1980 to 2013
df_least5 = df_can.tail(5)
# transpose the dataframe
df_least5 = df_least5[years].transpose() 
df_least5.head()
df_least5.index = df_least5.index.map(int) 
df_least5.plot(kind='area', alpha=0.45, figsize=(20, 10)) 
plt.title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
#Create an unstacked area plot of the 5 countries that contributed the least to immigration to Canada from 1980 to 2013
df_least5 = df_can.tail(5)

df_least5 = df_least5[years].transpose() 
df_least5.head()

df_least5.index = df_least5.index.map(int)
ax = df_least5.plot(kind='area', alpha=0.55, stacked=False, figsize=(20, 10))

ax.set_title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
# Analyze the effect of crisis in Argentina, and compare Argentina's immigration to that of it's neighbor Brazil. Use a bubble plot to plot the immigration from Brazil and Argentina for the years 1980 - 2013[includes normalisation]
```

```python
df_can_t=df_can[years].transpose()
df_can_t.index=map(int,df_can_t.index)
df_can_t.index.name='Year'

df_can_t.reset_index(inplace=True)

df_can_t.head()
df_can_t['Brazil']
```

```python
norm_brazil=(df_can_t['Brazil']-df_can_t['Brazil'].min())/(df_can_t['Brazil'].max()-df_can_t['Brazil'].min())

norm_argentina=(df_can_t['Argentina']-df_can_t['Argentina'].min())/(df_can_t['Argentina'].max()-df_can_t['Argentina'].min())
```

```python
ax0=df_can_t.plot(kind='scatter',
                 x='Year',
                 y='Brazil',
                 figsize=(14,8),
                 alpha=0.5,
                 color='blue',
                 s=norm_brazil*2000+10,
                 xlim=(1975,2015))

ax1=df_can_t.plot(kind='scatter',
                 x='Year',
                 y='Brazil',
                 figsize=(14,8),
                 alpha=0.5,
                 color='green',
                 s=norm_argentina*2000+10,
                 ax=ax0)
```

kdeplot

Plot univariate or bivariate distributions using kernel density estimation.

ecdfplot

distplot

DEPRECATED: Flexibly plot a univariate distribution of observations.

```python
anscombe=sns.load_dataset("anscombe")
#anscombe.head()
sns.set()
sns.lmplot(x="x",y="y",data=anscombe.query("dataset=='I'"),ci=None,scatter_kws={"s":80})
```

```python
sns.lmplot(x="x",y="y",data=anscombe.query("dataset=='II'"),order=2,ci=None,scatter_kws={"s":80})
```

```python
sns.lmplot(x="x",y="y",data=anscombe.query("dataset=='III'"),ci=None,robust=True,scatter_kws={"s":80});
```
